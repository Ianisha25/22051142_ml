```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the datasets
linearX_path = "linearX.csv"
linearY_path = "linearY.csv"

# Reading the data
X = pd.read_csv(linearX_path, header=None).values.flatten()
y = pd.read_csv(linearY_path, header=None).values.flatten()

# Display the first few entries for verification
X[:5], y[:5]
```

```python
X_mean = np.mean(X)
X_std = np.std(X)
X_normalized = (X - X_mean) / X_std

# Visualize the dataset
plt.figure(figsize=(8, 6))
plt.scatter(X_normalized, y, color='purple', label='Data Points')
plt.title('Scatter Plot of Normalized Predictor vs Dependent Variable')
plt.xlabel('Normalized Predictor (X)')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gradient Descent Function for Linear Regression
def gradient_descent(X, y, learning_rate=0.5, num_iterations=50):
    m = len(y)  # Number of training examples
    theta_0, theta_1 = 0, 0  # Initial parameters
    cost_history = []

    for iteration in range(num_iterations):
        # Hypothesis
        predictions = theta_0 + theta_1 * X
        errors = predictions - y

        # Cost function
        cost = (1 / (2 * m)) * np.sum(errors ** 2)
        cost_history.append(cost)

        # Gradients
        gradient_theta_0 = (1 / m) * np.sum(errors)
        gradient_theta_1 = (1 / m) * np.sum(errors * X)

        # Update parameters
        theta_0 -= learning_rate * gradient_theta_0
        theta_1 -= learning_rate * gradient_theta_1

    return theta_0, theta_1, cost_history

# Apply gradient descent with learning rate 0.5
theta_0, theta_1, cost_history = gradient_descent(X_normalized, y, learning_rate=0.5)

# Print the final parameters and plot the cost function
theta_0, theta_1
```

```python
# Plot cost function vs iteration
plt.figure(figsize=(8, 6))
plt.plot(range(1, 51), cost_history, marker='x', color='red', label='Cost Function')
plt.title('Cost Function vs Iteration (Learning Rate = 0.5)')
plt.xlabel('Iteration')
plt.ylabel('Cost Function Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Plot data points
plt.scatter(X_normalized, y, color='pink', label='Data Points')

# Plot the fitted straight line
fitted_line = theta_0 + theta_1 * X_normalized
plt.plot(X_normalized, fitted_line, color='purple', label='Fitted Line')

# Labels and title
plt.title('Dataset and Fitted Line (Linear Regression)')
plt.xlabel('Normalized Predictor (X)')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def plot_cost_vs_iteration_for_learning_rates(learning_rates, num_iterations=50):
    plt.figure(figsize=(12, 8))

    for lr in learning_rates:
        # Run gradient descent
        _, _, cost_history = gradient_descent(X_normalized, y, learning_rate=lr, num_iterations=num_iterations)
        
        # Plot cost function vs iteration
        plt.plot(range(1, num_iterations + 1), cost_history, marker='o', label=f'Learning Rate = {lr}')
    
    # Plot settings
    plt.title('Cost Function vs Iteration for Different Learning Rates')
    plt.xlabel('Iteration')
    plt.ylabel('Cost Function Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Test and plot for learning rates 0.005, 0.5, and 5
plot_cost_vs_iteration_for_learning_rates([0.005, 0.5, 5])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Stochastic Gradient Descent Function
def stochastic_gradient_descent(X, y, learning_rate=0.1, num_iterations=50):
    m = len(y)
    theta_0, theta_1 = 0, 0
    cost_history = []

    for iteration in range(num_iterations):
        total_cost = 0
        for i in range(m):
            prediction = theta_0 + theta_1 * X[i]
            error = prediction - y[i]
            
            # Update parameters
            theta_0 -= learning_rate * error
            theta_1 -= learning_rate * error * X[i]
            
            # Cost for each point
            total_cost += (error ** 2)
        cost_history.append(total_cost / (2 * m))
    
    return theta_0, theta_1, cost_history

# Mini-Batch Gradient Descent Function
def mini_batch_gradient_descent(X, y, learning_rate=0.1, num_iterations=50, batch_size=10):
    m = len(y)
    theta_0, theta_1 = 0, 0
    cost_history = []

    for iteration in range(num_iterations):
        indices = np.random.permutation(m)
        X_shuffled = X[indices]
        y_shuffled = y[indices]

        total_cost = 0
        for i in range(0, m, batch_size):
            X_batch = X_shuffled[i:i+batch_size]
            y_batch = y_shuffled[i:i+batch_size]

            predictions = theta_0 + theta_1 * X_batch
            errors = predictions - y_batch
            
            # Gradients
            gradient_theta_0 = np.sum(errors) / len(X_batch)
            gradient_theta_1 = np.sum(errors * X_batch) / len(X_batch)

            # Update parameters
            theta_0 -= learning_rate * gradient_theta_0
            theta_1 -= learning_rate * gradient_theta_1

            # Cost for the batch
            total_cost += np.sum(errors ** 2)
        
        cost_history.append(total_cost / (2 * m))
    
    return theta_0, theta_1, cost_history

# Plotting comparison of methods
def plot_comparisons():
    # Batch Gradient Descent
    theta_0_batch, theta_1_batch, cost_history_batch = gradient_descent(X_normalized, y, learning_rate=0.1)
    
    # Stochastic Gradient Descent
    theta_0_sgd, theta_1_sgd, cost_history_sgd = stochastic_gradient_descent(X_normalized, y, learning_rate=0.1)
    
    # Mini-Batch Gradient Descent
    theta_0_mini, theta_1_mini, cost_history_mini = mini_batch_gradient_descent(X_normalized, y, learning_rate=0.1)

    # Plotting cost vs iteration
    plt.figure(figsize=(10, 6))
    plt.plot(cost_history_batch, label='Batch Gradient Descent')
    plt.plot(cost_history_sgd, label='Stochastic Gradient Descent')
    plt.plot(cost_history_mini, label='Mini-Batch Gradient Descent')
    plt.title('Cost Function vs Iterations for Different Methods')
    plt.xlabel('Iterations')
    plt.ylabel('Cost Function Value')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the comparison plot function
plot_comparisons()
```
